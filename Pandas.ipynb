```python
import pandas as pd

df1 = pd.DataFrame({
    'emp_id': [1, 2, 3],
    'name': ['Alice', 'Bob', 'Charlie']
})

df2 = pd.DataFrame({
    'emp_id': [2, 3, 4],
    'salary': [50000, 60000, 70000]
})

result = pd.merge(df1, df2, on='emp_id', how='inner')
print(result)
```

```python
df3 = pd.DataFrame({
    'name': ['Alice', 'Bob'],
    'age': [25, 30]
})

df4 = pd.DataFrame({
    'name': ['Charlie', 'David'],
    'age': [35, 40]
})

result = pd.concat([df3, df4], axis=0)
print(result)
```

```python
# Feature engineering: apply(), map(), lambda
df['sex'] = df['sex'].map({'Male': 0, 'Female': 1})
df
```

```python
df['size'] = df['size'].apply(lambda x: x * 2)
df
```

```python
df['smoker'] = df['smoker'].apply(lambda x: len(x))
df
```

```python
# Date - Time Operations
df = pd.DataFrame({
    'order_date': ['2023-01-01', '2023-03-15', '2023-07-20']
})
df['order_date'] = pd.to_datetime(df['order_date'])

df['year'] = df['order_date'].dt.year
df['is_weekend'] = df['order_date'].dt.weekday >= 5
df['delivery_date'] = df['order_date'] + pd.Timedelta(days=7)
df
```
